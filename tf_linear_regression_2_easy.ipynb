# TF Linear Regression

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/content/sample_data/california_housing_train.csv")
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
trainx = np.asarray(df[[ 'housing_median_age', 'total_rooms',
       'total_bedrooms', 'population', 'households', 'median_income']])

trainy = np.asarray(df['median_house_value'])
trainx[:5]
```

```python
#define linear regression model

model = Sequential()

model.add(Dense(1, input_dim = len(trainx[0,:]) , activation = 'linear'))

model.compile(loss = 'mean_squared_error', optimizer = "rmsprop", metrics = [tf.keras.metrics.MeanAbsolutePercentageError() ])

```

```python
model.fit(x= trainx , y= trainy , epochs = 1000 , verbose = 2)
```

```python
#now test model with test data
test_df = pd.read_csv("/content/sample_data/california_housing_test.csv")
test_df.head()
```

```python
testx = np.asarray(test_df[[ 'housing_median_age', 'total_rooms',
       'total_bedrooms', 'population', 'households', 'median_income']])

testy = np.asarray(test_df['median_house_value'])
trainx[:5]
```

```python
#make prediction
pred  =  model.predict(testx)
pred[:10]
```
