# TF Logistic Regression

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/content/ChurnData.csv")
df.head()
```

```python
df.drop(['tenure','voice','ed','employ','equip','callcard','pager','internet','callwait','confer','ebill'],inplace =True, axis = 1)
df.columns
```

```python
X = np.asarray(df[['age', 'address', 'income', 'wireless' , 'wiremon', 'longten', 'tollten', 'cardten',
                    'loglong', 'logtoll', 'lninc', 'custcat' ]])

Y = np.asarray(df['churn'])
X[:5]
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X = scaler.fit_transform(X)

X[:5]
```

```python
y_uniq = set(Y)
y_uniq
```

```python
# split data
from sklearn.model_selection import train_test_split
trainx ,testx, trainy ,testy = train_test_split(X,Y, test_size=.2 , random_state=42)
```

```python
# define logistic regression model
model = Sequential()

model.add(Dense(  1, input_dim = len(X[0,:]) , activation = 'sigmoid'))

model.compile(  loss = 'binary_crossentropy' ,  optimizer='rmsprop', metrics = ['accuracy']  )

```

```python
model.fit(  x=X,   y=Y ,  epochs = 500 ,   verbose = 1 )
```

```python
y_pred  = model.predict(testx)
```
